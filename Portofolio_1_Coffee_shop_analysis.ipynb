# A. Join Table

```python
join_table = pd.read_sql_query('''select a.product_id, product_category, unit_of_measure, current_wholesale_price, 
                                  promo_yn, new_product_yn, transaction_id, transaction_date, transaction_time, 
                                  a.sales_outlet_id, customer_id, instore_yn, quantity, unit_price, promo_item_yn,
                                  sales_outlet_type, store_city,store_longitude, store_latitude, Neighorhood
                                  from 'sales reciepts' a
                                  left join product b 
                                  on a.product_id = b.product_id
                                  left join 'sales outlet' c
                                  on a.sales_outlet_id = c.sales_outlet_id
                                  ''', conn)
```
